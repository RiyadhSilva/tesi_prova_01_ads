## Baixe o título das 5 notícias da página inicial;

```python
import requests
import requests_cache

requests_cache.install_cache()
```

```python
def download(url, num_retries=2): 
    print('Downloading:', url) 
    page = None 
    try: 
        response = requests.get(url) 
        page = response.text 
        if response.status_code >= 400: 
            print('Download error:', response.text) 
            if num_retries and 500 <= response.status_code < 600: 
                return download(url, num_retries - 1) 
    except requests.exceptions.RequestException as e: 
                print('Download error:', e.reason) 
    return page
```

```python
page = download('http://www.tce.pi.gov.br/')
```

```python
from bs4 import BeautifulSoup
url = 'http://www.tce.pi.gov.br/'
html = download(url)
soup = BeautifulSoup(html, 'html.parser')
fixed_html = soup.prettify()
print(fixed_html)
```

```python
noticias = soup.findAll('a', class_='latestnews')
```

```python
cont = 1;
lista_noticia = []
lista_links = []
for noticia in noticias:
    print(cont, "ª Notícia: ", noticia.getText())
    lista_noticia.append(noticia.getText())
    lista_links.append(noticia['href'])
    cont = cont + 1
```

## Forneça uma opção um menu com escolha de 1 a 6 correspondendo às 5 matérias e uma sexta opção de sair; Caso o usuário escolha uma das opções, o texto da mensagem deve ser baixado e exibido. ; Após exibir a matéria, deve-se da a opção para exibir o menu novamente.

```python
dado = True;
while dado:
    print('------ Bem vindo ao NotiPI -----')
    print('Entre com um numero de 1 a 5 para acessar alguma materia')
    cont = 1;
    for noticia in lista_noticia:
        print(cont, " - ", noticia)
        cont = cont+1
    print('Entre com 6 para sair da aplicação')
    entrada = input('Entre com um número: ')
    if entrada == "6":
        dado = False
    else:
        print("\n")
        print("Título da notícia: ", lista_noticia[int(entrada)-1])
        link = lista_links[int(entrada)-1]
        print("Link da notícia: ", link)
        html = download(link)
        soup = BeautifulSoup(html, 'html.parser')
        fixed_html = soup.prettify()
        conteudos = soup.select('div.the_post.post_content')
        for conteudo in conteudos:
            print(conteudo.getText())
        print("\n\n")
```
